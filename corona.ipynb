```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sas
```

```python
df = pd.read_excel("corona.xlsx")
```

```python
df
```

```python
df.info()
```

```python
my_color = ['orange', 'blue', 'red', 'green', 'black']

# Prepare the DataFrame with top 5 states
top_states = df.groupby(['State'], as_index=False)['Total _Cases'].sum() \
                .sort_values(by='Total _Cases', ascending=False) \
                .head(5)

# Set the figure size using Matplotlib's rcParams
plt.figure(figsize=(20, 10))

# Create the bar plot using Seaborn
sas.barplot(x='State', y='Total _Cases', data=top_states, palette=my_color)

# Add titles and labels (optional)
plt.title('Top 5 States by Total Cases')
plt.xlabel('State')
plt.ylabel('Total Cases')

# Show the plot
plt.show()
```

```python
my_color = ['orange', 'blue', 'red', 'green', 'black']
top_states = df.groupby(['State'], as_index=False)['Vacasination _done'].sum() \
                .sort_values(by='Vacasination _done', ascending=False) \
                .head(5)
plt.figure(figsize=(20, 10))
sas.barplot(x='State', y='Vacasination _done', data=top_states, hue = 'State')
plt.title('Top 5 States by Vacasination _done')
plt.xlabel('State')
plt.ylabel('Vacasination _done')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sizes = [df['Male'].sum(), df['Female'].sum()] 
labels = ['Male', 'Female']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sizes = [df['Covisheld_ Doses'].sum(), df['Covaxin _Dose'].sum()] 
labels = ['Covisheld_ Doses', 'Covaxin _Dose']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(40, 18))

plt.plot(df['State'], df['Dose 1'], marker='o', linestyle='-', color='b', label='Dose 1')
plt.plot(df['State'], df['Dose 2'], marker='s', linestyle='--', color='r', label='Dose 2')
plt.title('Line Graph of Dose 1 and Dose 2')
plt.xlabel('State')  # Adjust based on what 'State' represents
plt.ylabel('Dose Amount')
plt.legend()

# Add grid lines (optional)
plt.grid(True)

# Show the plot
plt.show()
```
